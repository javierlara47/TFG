PARSEO DE LA COLUMNA VALUE

```python
#La columna value contiene valores en formato str, el codigo lo que hace es eliminar cualquier caracter no numerico, extraer los digitos
#y convierte la columna resultante en valores numericos de tipo FLOAT
#El resultado se almacena en una nueva columna llamada value_numeric

def convertir_valor(valor):
    if isinstance(valor, str):  # Solo procesar si es un string
        valor = valor.replace("€", "").replace(",", ".")  # Quitar € y cambiar , por .
        if "M" in valor:
            return int(float(valor.replace("M", "")) * 1_000_000)  # Convertir millones a enteros
        elif "K" in valor:
            return int(float(valor.replace("K", "")) * 1_000)  # Convertir miles a enteros
        else:
            return int(valor)  # Si no tiene M ni K, convertir directamente
    return valor  # Si ya es un número, devolverlo igual
```

```python
#Cambiar el nombre de la columna de inglés a español
df_futbol.rename(columns={'Value':'valor_de_mercado'},inplace=True)
```

```python
#Ordenar el DataFrame en función de la columna valor_de_mercado, de manera descendente
df_ordenado=df_futbol.sort_values(by='valor_de_mercado',ascending=False)

#Muestra por pantalla las columnas de nombre y valor_de_mercado solo
df_ordenado[['name','valor_de_mercado']]
```

```python
# Función para limpiar y convertir valores a enteros
def convertir_valor(valor):
    if isinstance(valor, str):  # Verificar si es string
        valor = valor.replace("€", "").replace(",", ".")  # Quitar símbolo €
        if "M" in valor:
            return int(float(valor.replace("M", "")) * 1_000_000)  # Convertir millones a enteros
        elif "K" in valor:
            return int(float(valor.replace("K", "")) * 1_000)  # Convertir miles a enteros
        else:
            return int(valor)  # Si no tiene M ni K, convertir directamente
    return valor  # Si ya es número, devolverlo igual

# Aplicar la conversión a la columna
df_futbol["valor_de_mercado"] = df_futbol["valor_de_mercado"].apply(convertir_valor)

# Verificar el tipo de dato después de la conversión
print(df_futbol["valor_de_mercado"].dtype)  # Debe mostrar int64
```

ESTADISTICA DESCRIPTIVA

```python
#Configurar el formato de visualizacion para mostrar numeros completos
pd.set_option('display.float_format',lambda x: '%.3f' % x)

df_futbol['valor_de_mercado'].describe()
```

```python
import matplotlib.pyplot as plt

#Ajustar la escala del eje Y para mostrar numeros completos
plt.ticklabel_format(style='plain', axis='y')

#Boxplot con matplotlib
plt.boxplot(df_futbol['valor_de_mercado'])
plt.title('Boxplot del valor de mercado')
plt.show()

#Histograma con matplotlib
plt.hist(df_futbol['valor_de_mercado'], bins=100)
plt.title('Histograma de valor de mercado')
plt.show()
```

```python
#Calcular el rango intercuartilico
Q1 = df_futbol['valor_de_mercado'].quantile(0.25)
Q3 = df_futbol['valor_de_mercado'].quantile(0.75)
IQR = Q3-Q1

#Definir los limites para considerar outliers
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR

#Filtrar el DataFrame para mantener solo los valores dentro de los limites
df_filtrado = df_futbol[(df_futbol['valor_de_mercado'] >= lower_bound) & (df_futbol['valor_de_mercado'] <= upper_bound)]
```
