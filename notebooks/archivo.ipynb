PARSEO DE LA COLUMNA VALUE

```python
#La columna value contiene valores en formato str, el codigo lo que hace es eliminar cualquier caracter no numerico, extraer los digitos
#y convierte la columna resultante en valores numericos de tipo FLOAT
#El resultado se almacena en una nueva columna llamada value_numeric

def convertir_valor(valor):
    if isinstance(valor, str):  # Solo procesar si es un string
        valor = valor.replace("€", "").replace(",", ".")  # Quitar € y cambiar , por .
        if "M" in valor:
            return int(float(valor.replace("M", "")) * 1_000_000)  # Convertir millones a enteros
        elif "K" in valor:
            return int(float(valor.replace("K", "")) * 1_000)  # Convertir miles a enteros
        else:
            return int(valor)  # Si no tiene M ni K, convertir directamente
    return valor  # Si ya es un número, devolverlo igual
```

1.1 VARIANZA

    -La varianza es una medida de la dispersión o extensión de un conjunto de datos
    -Indica cuanto se alejan los valores individuales del valor medio (promedio) del conjunto
    -Si los datos están muy dispersos, la varianza será alta; si estan agrupados al rededor de la media, la varianza es baja.
    -Se calcula sumando los cuadrados de las diferencias entre cada valor y la media, y luego dividiendo por el numero total de 

```python
#Varianza
df_filtrado['valor_de_mercado'].var()
```

1.2 Desviación Estándar

    -La desviación estándar es simplemente la raíz cuadrada de la varianza
    -Proporciona una medida de dispersión en la misma escala que los datos originales
    -Es una medida más facil de interpretar porque esta en las mismas unidades que los datos
    -Si la desviación estandar es pequeña, significa que los datos tienden a estar cerca del promedio; si es grande, significa que los dtos estan mas dispersos

```python
df_filtrado['valor_de_mercado'].std()
```

Correlacion

    La correlación entre dos variables es una medida estadística que describe la relación estadistica y la dependencia entre ellas. En otras palabras, la correlación indica como cambia una variable en relacion a la otra.

    La correlacion mide la relacion lineal entre dos variables. Este coeficiente varía entre -1 y 1:
        
        -Si es 1, significa una correlación positiva perfecta.
        -Si es -1, indica una correlacion negativa perfecta.
        -Si es 0, no hay correlacion lineal
