PARSEO DE LA COLUMNA VALUE

```python
#La columna value contiene valores en formato str, el codigo lo que hace es eliminar cualquier caracter no numerico, extraer los digitos
#y convierte la columna resultante en valores numericos de tipo FLOAT
#El resultado se almacena en una nueva columna llamada value_numeric

def convertir_valor(valor):
    if isinstance(valor, str):  # Solo procesar si es un string
        valor = valor.replace("€", "").replace(",", ".")  # Quitar € y cambiar , por .
        if "M" in valor:
            return int(float(valor.replace("M", "")) * 1_000_000)  # Convertir millones a enteros
        elif "K" in valor:
            return int(float(valor.replace("K", "")) * 1_000)  # Convertir miles a enteros
        else:
            return int(valor)  # Si no tiene M ni K, convertir directamente
    return valor  # Si ya es un número, devolverlo igual
```

```python
df_futbol_sin_nulos['Value'] = df_futbol_sin_nulos['Value'].apply(convertir_valor)
```
