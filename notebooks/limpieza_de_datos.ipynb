```python
import re

#Lista con todas las posibles posiciones de los jugadores, usada para identificar y extrar las posiciones de cada uno
#de los jugadores
posiciones = ['GK', 'LB', 'LWB', 'RB', 'RWB', 'CB', 'CAM', 'CM', 'CDM', 'LW', 'RW' ,'LM', 'RM', 'LF', 'RF', 'CF', 'ST']
def split_name_position(fila):
    fila = re.sub(r'\d+', '', fila)  #Elimina todos lo numeros
    palabra = fila.split()  #Dividir la cadena en palabras individuales (espacios)
    nombre_completo = [part for part in palabra if part not in posiciones]  #Guardar las palabras que no son una posicion
    posicion = [part for part in palabra if part in posiciones]  #Guardar las palabras que sí son una posicion
    

    nombre = ' '.join(nombre_completo) #Une las palabras para construir el nombre del futbolista
    pos = ', '.join(posicion) #Une las palabras (separadas por comas si es necesario) para contruir la posicion/es
    return pd.Series([nombre, pos])

#Eliminamos la columna name, y añadimos dos nuevas columnas, llamadas "player_name" y "position"
df[['player_name', 'position']] = df['name'].apply(split_name_position)
df.drop('name', axis=1, inplace=True)
```

```python
#Divido la columna en equipo y años de contrato
df[['team', 'contract']] = df['team_&_contract'].str.split('\n', expand=True)

#Creo una columna que almacena el año de inicio del contrato
df['contract_start'] = df['contract'].str.split(' ~ ').str[0]
#Creo una columna que almacena el año de finalizacion del contrato
df['contract_end'] = df['contract'].str.split(' ~ ').str[-1]

#En la columna contract_end puede haber texto adicional, por lo tanto se limpia dicha columna.
#En el caso de que la columna tenga texto adicional (es decir que sea texto), me quedo con los ultimos 4 caracteres
df['contract_end'] = df['contract_end'].apply(lambda x: x if x.isdigit() else x[-4:])
df.drop(['team_&_contract', 'contract'], axis=1, inplace=True)

df.head()
```

```python
#Con este patrón, se buscarán valores que sean años de 4 dígitos y se ignorará cualquier otro texto 
#(como "loan", "Exp. in 2025", etc)
patron = r'^\d{4}$'

#Para cada una de las columnas, convierte el valor a texto y con el patron definido anteriormente, comprueba si coincide.
#Si coincide, axtrae el año y lo mantiene en la columna
#Si no coincide, reemplaza el valor por una cadena vacia
df['contract_start'] = df['contract_start'].apply(lambda x: re.match(patron, str(x)).group() 
                                                  if re.match(patron, str(x)) else '')
df['contract_end'] = df['contract_end'].apply(lambda x: re.match(patron, str(x)).group() 
                                              if re.match(patron, str(x)) else '')

#Teniendo las columnas como valores numericos, convierte los valores vacios a valores nulos (pd.to_numeric)
#Y despues los sustituye por valor 0, para asegurar que la columna sea numerica (.fillna(0))
df['contract_start'] = pd.to_numeric(df['contract_start'], errors='coerce').fillna(0).astype(int)
df['contract_end'] = pd.to_numeric(df['contract_end'], errors='coerce').fillna(0).astype(int)
```

```python
#Limpieza para las columnas 'value', 'wage' y 'release_clause'
def convertir_valores(valores):
    if 'K' in valores: #Si el valor contiene "K"
            #Convierte el valor en miles, multiplicandolo por 1000
            return float(valores.replace('€','').replace('K',''))*1000
    elif 'M' in valores: #Si el valor contiene "M"
            #Convierte el valor en millones, multiplicandolo por 1000000
            return float(valores.replace('€','').replace('M',''))*1000000
    else:
            #Si el valor no contiene ni "K" ni "M", elimina "€"
            return float(valores.replace('€',''))
```

```python
#Aplicar la funcion anterior a las columnas
df['value'] = df['value'].apply(convertir_valores)
df['wage'] = df['wage'].apply(convertir_valores)
df['release_clause'] = df['release_clause'].apply(convertir_valores)
```

```python
import pandas as pd
import re

# ---------- columnas que deseas limpiar ----------
columns_to_clean = [
    "age", "overall_rating", "potential",
    "crossing", "finishing", "heading_accuracy", "short_passing", "volleys",
    "dribbling", "curve", "fk_accuracy", "long_passing", "ball_control",
    "acceleration", "sprint_speed", "agility", "reactions", "balance",
    "shot_power", "jumping", "stamina", "strength", "long_shots",
    "aggression", "interceptions", "att._position", "vision",
    "penalties", "composure", "defensive_awareness", "standing_tackle",
    "sliding_tackle", "gk_diving", "gk_handling", "gk_kicking",
    "gk_positioning", "gk_reflexes", "base_stats", "international_reputation"
]

# ---------- función: extrae la parte numérica antes de +n o -n ----------
def strip_plus_minus(series: pd.Series) -> pd.Series:
    """
    '83+1' → 83.0
    '79-1' → 79.0
    '72'   → 72.0
    Lanza ValueError si no hay número al principio.
    """
    # 1) fuerza a str, 2) extrae el número inicial (regex), 3) convierte a float
    cleaned = (
        series.astype(str)
              .str.extract(r"^(\d+)", expand=False)   # solo dígitos iniciales
    )
    return pd.to_numeric(cleaned, errors="raise").astype(float)

# ---------- aplicar limpieza ----------
for col in columns_to_clean:
    if col in df.columns:
        df[col] = strip_plus_minus(df[col])

print("✅ Limpieza completada: eliminados '+n' y '-n', convertido a float")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular el rango intercuartílico (IQR)
Q1 = df['value'].quantile(0.25)  # Primer cuartil (25%)
Q3 = df['value'].quantile(0.75)  # Tercer cuartil (75%)
IQR = Q3 - Q1  # Rango intercuartílico

# Definir los límites para considerar *outliers*
lower_bound = Q1 - 1.5 * IQR  # Límite inferior
upper_bound = Q3 + 1.5 * IQR  # Límite superior

# Filtrar el DataFrame para mantener solo los valores dentro de los límites
df_filtrado = df[(df['value'] >= lower_bound) & (df['value'] <= upper_bound)]

# Generar el boxplot con los valores filtrados
plt.figure(figsize=(7,5))
sns.boxplot(x=df_filtrado['value'])
plt.xlabel('Valor de los jugadores sin outliers')
plt.title('Distribución del valor de los jugadores tras eliminar valores atípicos')
plt.show()

```
