```python
import re

#Lista con todas las posibles posiciones de los jugadores, usada para identificar y extrar las posiciones de cada uno
#de los jugadores
posiciones = ['GK', 'LB', 'LWB', 'RB', 'RWB', 'CB', 'CAM', 'CM', 'CDM', 'LW', 'RW' ,'LM', 'RM', 'LF', 'RF', 'CF', 'ST']
def split_name_position(fila):
    fila = re.sub(r'\d+', '', fila)  #Elimina todos lo numeros
    palabra = fila.split()  #Dividir la cadena en palabras individuales (espacios)
    nombre_completo = [part for part in palabra if part not in posiciones]  #Guardar las palabras que no son una posicion
    posicion = [part for part in palabra if part in posiciones]  #Guardar las palabras que sí son una posicion
    

    nombre = ' '.join(nombre_completo) #Une las palabras para construir el nombre del futbolista
    pos = ', '.join(posicion) #Une las palabras (separadas por comas si es necesario) para contruir la posicion/es
    return pd.Series([nombre, pos])

#Eliminamos la columna name, y añadimos dos nuevas columnas, llamadas "player_name" y "position"
df[['player_name', 'position']] = df['name'].apply(split_name_position)
df.drop('name', axis=1, inplace=True)
```

Como se podía observar en el dataset original, la columna "name" contenía tanto el nombre del jugador como su/s posicion/es. Si una columna contiene mas de un tipo de informacion, es difícil de manejar en análisis estadísticos y modelos de Machine Leanning. Por lo tanto, he optado por separar la columna "name" en dos columnas específicas:

    -"player_name" -> Contiene el nombre del jugador
    -"position" -> Contiene la/s posicion/es en las que cada jugador puede jugar.

Una vez hecha la limpieza anterior y separar las columnas de los años de contrato, dichas columnas son de tipo object aún, ya que hay datos de tipo "string" como "loan". Por lo tanto, estas dos columnas necesitan una limpieza que haga que sean de tipo "int"

```python
#Con este patrón, se buscarán valores que sean años de 4 dígitos y se ignorará cualquier otro texto 
#(como "loan", "Exp. in 2025", etc)
patron = r'^\d{4}$'

#Para cada una de las columnas, convierte el valor a texto y con el patron definido anteriormente, comprueba si coincide.
#Si coincide, axtrae el año y lo mantiene en la columna
#Si no coincide, reemplaza el valor por una cadena vacia
df['contract_start'] = df['contract_start'].apply(lambda x: re.match(patron, str(x)).group() 
                                                  if re.match(patron, str(x)) else '')
df['contract_end'] = df['contract_end'].apply(lambda x: re.match(patron, str(x)).group() 
                                              if re.match(patron, str(x)) else '')

#Teniendo las columnas como valores numericos, convierte los valores vacios a valores nulos (pd.to_numeric)
#Y despues los sustituye por valor 0, para asegurar que la columna sea numerica (.fillna(0))
df['contract_start'] = pd.to_numeric(df['contract_start'], errors='coerce').fillna(0).astype(int)
df['contract_end'] = pd.to_numeric(df['contract_end'], errors='coerce').fillna(0).astype(int)
```

```python
def convertir_valores(valores):
    if 'K' in valores: #Si el valor contiene "K"
            #Convierte el valor en miles, multiplicandolo por 1000
            return float(valores.replace('€','').replace('K',''))*1000
    elif 'M' in valores: #Si el valor contiene "M"
            #Convierte el valor en millones, multiplicandolo por 1000000
            return float(valores.replace('€','').replace('M',''))*1000000
    else:
            #Si el valor no contiene ni "K" ni "M", elimina "€"
            return float(valores.replace('€',''))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular el rango intercuartílico (IQR)
Q1 = df['value'].quantile(0.25)  # Primer cuartil (25%)
Q3 = df['value'].quantile(0.75)  # Tercer cuartil (75%)
IQR = Q3 - Q1  # Rango intercuartílico

# Definir los límites para considerar *outliers*
lower_bound = Q1 - 1.5 * IQR  # Límite inferior
upper_bound = Q3 + 1.5 * IQR  # Límite superior

# Filtrar el DataFrame para mantener solo los valores dentro de los límites
df_filtrado = df[(df['value'] >= lower_bound) & (df['value'] <= upper_bound)]

# Generar el boxplot con los valores filtrados
plt.figure(figsize=(16,5))
sns.boxplot(x=df_filtrado['value'])
plt.xlabel('Valor de los jugadores sin outliers')
plt.title('Distribución del valor de los jugadores tras eliminar valores atípicos')
plt.show()

```

En este mapa de calor (HeatMap), los colores representan la fuerza y dirección de correlación entre las variables numericas. Para entenderlo mejor, primero hay que conocer lo qué significa la escala de correlación que va de -1 a 1.

La correlación mide como dos variables están relacionadas, y esto se expresa en un rango que va de -1 a 1, donde:

    - (+1) (correlación positiva perfecta): Cuando una variable aumenta, la otra aumenta en la misma proporción.
    - (-1) (correlación negativa perfecta): Cuando una variable aumenta, la otra disminuye en la misma proporción.
    - 0 (sin correlacion): No hay relacion entre las dos variables.

Observando el gráfico, vemos como se hace uso de colores para representar el rango anteriormente explicado. Estos colores se interpretan de la siguiente forma:

    -Las zonas blancas indican relaciones fuertes entre variables.
    -Las zonas negras indican relaciones inversas, lo que sugiere que un aumento en una variable implica una reducción en otra.
    -Las zonas rojizas indican que dos variables no tienen una relación clara, por lo que no sería útil incluirlas juntas en un modelo predictivo.

Una vez sabemos esto, ya podemos detectar que variables son las que tienen una fuerte relación con el valor de mercado de los jugadores.

Calculo del número óptimo de clusters usando Silhouette

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


#Seleccionar las variables adecuadas para clustering
columns_for_clustering = [
    "crossing","heading_accuracy","acceleration",
    "ball_control", "fk_accuracy","dribbling", "short_passing", "long_passing",
    "finishing", "balance","jumping","shot_power", "long_shots","att._position" ,"vision",
    "composure", "aggression", "interceptions", "defensive_awareness",
    "gk_reflexes", "gk_positioning", "gk_diving","gk_handling","gk_kicking"
]

#Filtrar solo las columnas numéricas y asegurarse de que no haya valores categóricos
df_clustering = df[columns_for_clustering].copy()

#Calcular el coeficiente de Silhouette para distintos valores de clusters
range_n_clusters = list(range(2, 11))
silhouette_avg_scores = []

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(df_clustering)
    silhouette_avg = silhouette_score(df_clustering, cluster_labels)
    silhouette_avg_scores.append(silhouette_avg)

#Graficar los resultados
plt.figure(figsize=(8, 5))
plt.plot(range_n_clusters, silhouette_avg_scores, marker='o', linestyle='-', color='b')
plt.xlabel("Número de Clusters (k)")
plt.ylabel("Coeficiente de Silhouette")
plt.title("Cálculo del Número Óptimo de Clusters usando Silhouette")
plt.xticks(range_n_clusters)
plt.grid(True)
plt.show()

#Mostrar el número óptimo de clusters
optimal_k = range_n_clusters[np.argmax(silhouette_avg_scores)]
print(f"✅ El número óptimo de clusters según Silhouette es: {optimal_k}")
```

DUNN

```python
from sklearn.metrics import pairwise_distances

#Función para calcular el Índice de Dunn
def dunn_index(X, labels):
    """
    Calcula el Índice de Dunn para una partición de clustering.
    X: Datos
    labels: Etiquetas de los clusters asignados por K-Means
    """
    unique_clusters = np.unique(labels)
    cluster_distances = []
    cluster_diameters = []

    for i in unique_clusters:
        for j in unique_clusters:
            if i != j:
                # Distancia mínima entre puntos de diferentes clusters
                min_intercluster_distance = np.min(
                    pairwise_distances(X[labels == i], X[labels == j])
                )
                cluster_distances.append(min_intercluster_distance)

        # Diámetro del cluster (máxima distancia dentro del mismo cluster)
        intra_distances = pairwise_distances(X[labels == i])
        cluster_diameter = np.max(intra_distances)
        cluster_diameters.append(cluster_diameter)

    return np.min(cluster_distances) / np.max(cluster_diameters)

#Filtrar solo las columnas numéricas y asegurarse de que no haya valores categóricos
df_clustering = df[columns_for_clustering].copy()

#Calcular el Índice de Dunn para distintos valores de clusters
range_n_clusters = list(range(2, 11))  # Probar de 2 a 10 clusters
dunn_scores = []

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(df_clustering.to_numpy())  # Convertir a array numpy
    dunn_score = dunn_index(df_clustering.to_numpy(), cluster_labels)
    dunn_scores.append(dunn_score)

#Graficar los resultados
plt.figure(figsize=(8, 5))
plt.plot(range_n_clusters, dunn_scores, marker='o', linestyle='-', color='b')
plt.xlabel("Número de Clusters (k)")
plt.ylabel("Índice de Dunn")
plt.title("Cálculo del Número Óptimo de Clusters usando el Índice de Dunn")
plt.xticks(range_n_clusters)
plt.grid(True)
plt.show()

#Mostrar el número óptimo de clusters
optimal_k_dunn = range_n_clusters[np.argmax(dunn_scores)]
print(f"✅ El número óptimo de clusters según el Índice de Dunn es: {optimal_k_dunn}")
```

```python


variables1 = df_field_players[["heading_accuracy", "acceleration",
                                "ball_control","agility","dribbling", 
                                "short_passing", "long_passing", "finishing", 
                                "shot_power", "long_shots", "att._position", "vision", "composure", "interceptions",
                                "defensive_awareness","sliding_tackle","standing_tackle"]]


scaler = StandardScaler()
variables = scaler.fit_transform(variables1)

# Definir el rango de k
k_range = range(2, 11)

silhouette_scores_new = []

# Calcular el Silhouette Score para cada k
for k in k_range:
    kmeans_new = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_new.fit(variables)
    silhouette_scores_new.append(silhouette_score(variables, kmeans_new.labels_))

# Identificar el número óptimo de clusters basado en el Silhouette Score
optimal_k_silhouette = k_range[silhouette_scores_new.index(max(silhouette_scores_new))]

# Graficar los resultados
plt.figure(figsize=(8, 6))
plt.plot(k_range, silhouette_scores_new, marker='o', linestyle='-', color='b')
plt.title("Silhouette Score para diferentes valores de k")
plt.xlabel("Número de Clusters (k)")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()

# Imprimir el número óptimo de clusters
print(f"El número óptimo de clusters según el Silhouette Score es: {optimal_k_silhouette}")
```

```python
from sklearn.metrics import calinski_harabasz_score

calinski_scores_new = []

# Calcular el Calinski-Harabasz Score para cada k
for k in k_range:
    kmeans_new = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_new.fit(variables)
    calinski_scores_new.append(calinski_harabasz_score(variables, kmeans_new.labels_))

# Identificar el número óptimo de clusters basado en el Calinski-Harabasz Score
optimal_k_calinski = k_range[calinski_scores_new.index(max(calinski_scores_new))]

# Graficar los resultados
plt.figure(figsize=(8, 6))
plt.plot(k_range, calinski_scores_new, marker='o', linestyle='-', color='g')
plt.title("Calinski-Harabasz Score para diferentes valores de k")
plt.xlabel("Número de Clusters (k)")
plt.ylabel("Calinski-Harabasz Score")
plt.grid(True)
plt.show()

# Imprimir el número óptimo de clusters
print(f"El número óptimo de clusters según el Calinski-Harabasz Score es: {optimal_k_calinski}")
```

```python
from scipy.spatial.distance import cdist

dunn_scores_new = []

# Calcular el Dunn Index para cada k
for k in k_range:
    kmeans_new = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_new.fit(variables)
    
    # Calcular la distancia entre los centros de los clusters
    dist = cdist(kmeans_new.cluster_centers_, kmeans_new.cluster_centers_, 'euclidean')
    
    # Asegurarse de no dividir por cero
    dist = dist[dist > 0]  # Excluir la distancia cero
    if dist.size == 0:  # Si no hay distancias, asignar Dunn a un valor mínimo
        dunn_score = 0
    else:
        # Calcular el Dunn Index
        dunn_score = min(dist) / max(cdist(variables, kmeans_new.cluster_centers_, 'euclidean').min(axis=1))
    
    dunn_scores_new.append(dunn_score)

# Identificar el número óptimo de clusters basado en el Dunn Index
optimal_k_dunn = k_range[dunn_scores_new.index(max(dunn_scores_new))]

# Graficar los resultados
plt.figure(figsize=(8, 6))
plt.plot(k_range, dunn_scores_new, marker='o', linestyle='-', color='r')
plt.title("Dunn Index para diferentes valores de k")
plt.xlabel("Número de Clusters (k)")
plt.ylabel("Dunn Index")
plt.grid(True)

print(f"El número óptimo de clusters según el Dunn Index es: {optimal_k_dunn}")
```
