```python
import re

#Lista con todas las posibles posiciones de los jugadores, usada para identificar y extrar las posiciones de cada uno
#de los jugadores
posiciones = ['GK', 'LB', 'LWB', 'RB', 'RWB', 'CB', 'CAM', 'CM', 'CDM', 'LW', 'RW' ,'LM', 'RM', 'LF', 'RF', 'CF', 'ST']
def split_name_position(fila):
    fila = re.sub(r'\d+', '', fila)  #Elimina todos lo numeros
    palabra = fila.split()  #Dividir la cadena en palabras individuales (espacios)
    nombre_completo = [part for part in palabra if part not in posiciones]  #Guardar las palabras que no son una posicion
    posicion = [part for part in palabra if part in posiciones]  #Guardar las palabras que sí son una posicion
    

    nombre = ' '.join(nombre_completo) #Une las palabras para construir el nombre del futbolista
    pos = ', '.join(posicion) #Une las palabras (separadas por comas si es necesario) para contruir la posicion/es
    return pd.Series([nombre, pos])

#Eliminamos la columna name, y añadimos dos nuevas columnas, llamadas "player_name" y "position"
df[['player_name', 'position']] = df['name'].apply(split_name_position)
df.drop('name', axis=1, inplace=True)
```

Como se podía observar en el dataset original, la columna "name" contenía tanto el nombre del jugador como su/s posicion/es. Si una columna contiene mas de un tipo de informacion, es difícil de manejar en análisis estadísticos y modelos de Machine Leanning. Por lo tanto, he optado por separar la columna "name" en dos columnas específicas:

    -"player_name" -> Contiene el nombre del jugador
    -"position" -> Contiene la/s posicion/es en las que cada jugador puede jugar.

```python
#Divido la columna en equipo y años de contrato
df[['team', 'contract']] = df['team_&_contract'].str.split('\n', expand=True)

#Creo una columna que almacena el año de inicio del contrato
df['contract_start'] = df['contract'].str.split(' ~ ').str[0]
#Creo una columna que almacena el año de finalizacion del contrato
df['contract_end'] = df['contract'].str.split(' ~ ').str[-1]

#En la columna contract_end puede haber texto adicional, por lo tanto se limpia dicha columna.
#En el caso de que la columna tenga texto adicional (es decir que sea texto), me quedo con los ultimos 4 caracteres
df['contract_end'] = df['contract_end'].apply(lambda x: x if x.isdigit() else x[-4:])
df.drop(['team_&_contract', 'contract'], axis=1, inplace=True)

df.head()
```

Con esta limpieza en la columna team&contract, se consigue separar la información en columnas específicas para así, optimizar la organización del dataset. Antes, toda la columna era de tipo "String", con lo que era mucho más dificil el poder trabajar con los años de contrato, en el caso que queramos analizar dichos valores.

Una vez hecha la limpieza anterior y separar las columnas de los años de contrato, dichas columnas son de tipo object aún, ya que hay datos de tipo "string" como "loan". Por lo tanto, estas dos columnas necesitan una limpieza que haga que sean de tipo "int"

```python
#Con este patrón, se buscarán valores que sean años de 4 dígitos y se ignorará cualquier otro texto 
#(como "loan", "Exp. in 2025", etc)
patron = r'^\d{4}$'

#Para cada una de las columnas, convierte el valor a texto y con el patron definido anteriormente, comprueba si coincide.
#Si coincide, axtrae el año y lo mantiene en la columna
#Si no coincide, reemplaza el valor por una cadena vacia
df['contract_start'] = df['contract_start'].apply(lambda x: re.match(patron, str(x)).group() 
                                                  if re.match(patron, str(x)) else '')
df['contract_end'] = df['contract_end'].apply(lambda x: re.match(patron, str(x)).group() 
                                              if re.match(patron, str(x)) else '')

#Teniendo las columnas como valores numericos, convierte los valores vacios a valores nulos (pd.to_numeric)
#Y despues los sustituye por valor 0, para asegurar que la columna sea numerica (.fillna(0))
df['contract_start'] = pd.to_numeric(df['contract_start'], errors='coerce').fillna(0).astype(int)
df['contract_end'] = pd.to_numeric(df['contract_end'], errors='coerce').fillna(0).astype(int)
```

LIMPIEZA PARA LAS COLUMNAS VALUE, WAGE Y RELEASE_CLAUSE

```python
def convertir_valores(valores):
    if 'K' in valores: #Si el valor contiene "K"
            #Convierte el valor en miles, multiplicandolo por 1000
            return float(valores.replace('€','').replace('K',''))*1000
    elif 'M' in valores: #Si el valor contiene "M"
            #Convierte el valor en millones, multiplicandolo por 1000000
            return float(valores.replace('€','').replace('M',''))*1000000
    else:
            #Si el valor no contiene ni "K" ni "M", elimina "€"
            return float(valores.replace('€',''))
```

```python
#Aplicar la funcion anterior a las columnas value,wage y release_clause
df['value'] = df['value'].apply(convertir_valores)
df['wage'] = df['wage'].apply(convertir_valores)
df['release_clause'] = df['release_clause'].apply(convertir_valores)
```

Con esta representación generada, se puede ver una distribución clara de los valores de mercado de los jugadores del dataset.

    -La mediana (línea dentro de la caja) también se encuentra en la parte baja de la distribución, confirmando que los valores están altamente concentrados en un rango bajo.

Otra cosa a tener en cuenta de la representación, es la presencia de un gran numero de valores atípicos:

    -Se pueden observar numerosos valores atípicos (outliers) representados como puntos individuales a la derecha del diagrama.
    -Esto sugiere que existen jugadores con valores de mercado extremadamente altos en comparación con el resto.
    -Estos outliers pueden corresponder a futbolistas de clase mundial, cuyos precios de mercado son significativamente superiores al promedio.

Por lo tanto, se puede ver como hay una gran disparidad en la valoración de los jugadores. Hay jugadores que tienen un valor de mercado muy superior a la media, mientras que la mayoria tienen un valor muy inferior a los 25 millones de euros.

Este resultado sugiere que sería interesante profundizar en el análisis de qué factores influyen en la valoración de los jugadores, así como estudiar posibles oportunidades de mercado para encontrar jugadores infravalorados en función de sus estadísticas.

Para tratar estos outliers, vamos a trabajar con el rango intercuartilico.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular el rango intercuartílico (IQR)
Q1 = df['value'].quantile(0.25)  # Primer cuartil (25%)
Q3 = df['value'].quantile(0.75)  # Tercer cuartil (75%)
IQR = Q3 - Q1  # Rango intercuartílico

# Definir los límites para considerar *outliers*
lower_bound = Q1 - 1.5 * IQR  # Límite inferior
upper_bound = Q3 + 1.5 * IQR  # Límite superior

# Filtrar el DataFrame para mantener solo los valores dentro de los límites
df_filtrado = df[(df['value'] >= lower_bound) & (df['value'] <= upper_bound)]

# Generar el boxplot con los valores filtrados
plt.figure(figsize=(16,5))
sns.boxplot(x=df_filtrado['value'])
plt.xlabel('Valor de los jugadores sin outliers')
plt.title('Distribución del valor de los jugadores tras eliminar valores atípicos')
plt.show()

```

```python
#Histograma con matplotlib
plt.hist(df_filtrado['value'], bins=100)
plt.title('Histograma de valor de mercado')
plt.show()
```

Una vez calculado el rango intercuartilico y haber eliminado los valores extremos, vemos en los resultados graficos como esto ha hecho que los datos sean más homogéneos y fáciles de analizar. Esto favorece a la precisión del modelo de Machine Learning, ya que los outliers pueden afectar negativamente a los resultados, permitiendo así centrarse en el mercado de jugadores "realistas", excluyendo a los más caros que pueden cambiar la media.

Con la eliminación de estos valores atípicos, se ha conseguido una representación más fiel del mercado real, permitiendo indentificar que la mayoría  de los jugadores se encuentran en un rango de valores más homogéneo. Sin embargo, aún persisten algunos outliers, lo que confirma que en el fútbol existen jugadores con una valoración significativamente mayor al resto.

Otro apunte importante es que una vez realizado esta eliminación, la línea amarilla (mediana) se acerca mucho mas en la representación grafica al valor que tiene el percentil 50% cuando ejecutamos "df.describe()"
